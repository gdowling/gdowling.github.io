### Simple Linear Regression

#### Overview

If you were to only learn one machine learning method, that method should the linear regression. Linear regression calculates the linear relationship between a single dependent variable and one or more independent variables. Basically, when our independent variables increase or decrease by a certain amount how much will our dependent increase or decrease.

The equation for a linear regression can be expressed as:


Y = mX + b

#### Math

The idea of a linear regression is to find a line of best fit when predicting our dependent variable. This means that we want to find a line that goes through a series of points such that we minimize the sum of the squared residuals. In layman terms we want a line that minimizes the distance between our predictions and the actual values. We square this difference to weight the positive and negatives misses equally.

Y represents our dependent variable or the variable that we are trying to predict/explain. X represents our independent variable or the variable that will help in predicting/explaining Y. m represents the slope of the regression line which represents the effect that X has on Y (i.e. If X increases by one unit then Y will increase by m units). b represents a constant or the known y intercept, if X were to equal zero then Y would be equal to b. 

#### Key Principles

_Statistical Significance_

An important thing to note when running any linear regression analysis is to ensure that the relationship between your independent variables and dependent variable is statistically significant. What that means is to check the p-value for each of the independent variables to make sure it meets your threshold. The standard threshold is a 95% confidence interval and corresponds to a p-value that is less than 0.05. What this means is we are 95% certain that the independent variable is not equal to zero, thus having no relationship with our dependent variable.

_Coefficient of Determination_

When running a regression, you will want to know how explanatory your independent variables are. This is measured using the coefficient of determination, better known as R-Squared. R-Squared is the proportion of the variance in the dependent variable that is predictable from the independent variables. It is between 0 and 1, where 1 would mean that 100% of the dependent variable’s variance is explained by your independent variables. There is no magic number for what is a good or acceptable R-Squared instead you must use your own intuition of what R-Squared would be adequate given your data.
